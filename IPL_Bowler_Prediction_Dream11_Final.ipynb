```python
import pandas as pd
from pycaret.regression import setup, compare_models, automl, save_model

# Loading and preprocessing the dataset
data = pd.read_csv('Bowler_DataSet_IPL_Dream11Points.csv')
data['Match Date'] = pd.to_datetime(data['Match Date'])

# Function to calculate weighted recent form for a given player and match date
def calculate_weighted_recent_form(player_data, match_date, num_matches=5):
    weights = [5, 4, 3, 2, 1]  # Define your weights here
    player_matches = player_data[player_data['Match Date'] < match_date]
    recent_matches = player_matches.sort_values(by='Match Date', ascending=False).head(num_matches)
    if len(recent_matches) == 0:
        return 0
    weighted_wickets = sum(recent_matches['Wickets'] * weights[:len(recent_matches)]) / sum(weights[:len(recent_matches)])
    return weighted_wickets

# Function to calculate overall average performance
def calculate_overall_avg_performance(player_data):
    if len(player_data) == 0:
        return 0
    return player_data['Wickets'].mean()

# Applying the functions to create separate metrics
# Player's recent and overall performance
data['Recent Performance Player Bowler'] = data.apply(
    lambda row: calculate_weighted_recent_form(data[data['Bowler Name'] == row['Bowler Name']], row['Match Date']), axis=1)
data['Overall Performance Player Bowler'] = data.apply(
    lambda row: calculate_overall_avg_performance(data[data['Bowler Name'] == row['Bowler Name']]), axis=1)

# Player's recent and overall performance against each opponent
data['Recent Performance Opponent Bowler'] = data.apply(
    lambda row: calculate_weighted_recent_form(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Opponent Team'] == row['Opponent Team'])], row['Match Date']), axis=1)
data['Overall Performance Opponent Bowler'] = data.apply(
    lambda row: calculate_overall_avg_performance(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Opponent Team'] == row['Opponent Team'])]), axis=1)

# Player's recent and overall performance in each stadium
data['Recent Performance Stadium Bowler'] = data.apply(
    lambda row: calculate_weighted_recent_form(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Stadium Name'] == row['Stadium Name'])], row['Match Date']), axis=1)
data['Overall Performance Stadium Bowler'] = data.apply(
    lambda row: calculate_overall_avg_performance(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Stadium Name'] == row['Stadium Name'])]), axis=1)

# Filling NaN values
data.fillna({'Recent Performance Player Bowler': 0, 'Overall Performance Player Bowler': 0, 
             'Recent Performance Opponent Bowler': 0, 'Overall Performance Opponent Bowler': 0, 
             'Recent Performance Stadium Bowler': 0, 'Overall Performance Stadium Bowler': 0}, inplace=True)

# Saving the preprocessed data
data.to_csv('Preprocessed_Bowler_Data.csv', index=False)

# Loading the preprocessed data for further analysis
data = pd.read_csv('Preprocessed_Bowler_Data.csv')

# Defining predictors and target variable
predictors = [
    'Bowler Name', 'Wickets', 'Overs', 'Balls Bowled', 'Runs Given', 'Maidens', 
    'Economy', 'Stadium Name', 'Opponent Team', 'Innings Type',
    'Recent Performance Player Bowler', 'Overall Performance Player Bowler',
    'Recent Performance Opponent Bowler', 'Overall Performance Opponent Bowler',
    'Recent Performance Stadium Bowler', 'Overall Performance Stadium Bowler'
]
target = 'Dream11 Points'

# Splitting the data into training and testing sets
train_data = data.sample(frac=0.7, random_state=123)
test_data = data.drop(train_data.index)

# Setup PyCaret environment
setup_config = setup(data=train_data[predictors + [target]], target=target, test_data=test_data[predictors + [target]], 
                     session_id=123, normalize=True, transformation=True, remove_multicollinearity=True, 
                     multicollinearity_threshold=0.95, verbose=False)

# Comparing all models and select the best one
best_model = compare_models()

# Getting the best model automatically
final_model = automl()

# Saving the best model
save_model(final_model, 'final_bowler_prediction_model')

print(f"The best model based on R2 score is: {final_model}")


# Further analysis and model creation can be done here
```

```python
import pandas as pd
from pycaret.regression import setup, compare_models, automl, save_model

# Loading and preprocessing the dataset
data = pd.read_csv('bowler_data_set_IplPart2023_Dream11.csv')
data['Match Date'] = pd.to_datetime(data['Match Date'])

# Function to calculate weighted recent form for a given player and match date
def calculate_weighted_recent_form(player_data, match_date, num_matches=5):
    weights = [5, 4, 3, 2, 1]  # Define your weights here
    player_matches = player_data[player_data['Match Date'] < match_date]
    recent_matches = player_matches.sort_values(by='Match Date', ascending=False).head(num_matches)
    if len(recent_matches) == 0:
        return 0
    weighted_wickets = sum(recent_matches['Wickets'] * weights[:len(recent_matches)]) / sum(weights[:len(recent_matches)])
    return weighted_wickets

# Function to calculate overall average performance
def calculate_overall_avg_performance(player_data):
    if len(player_data) == 0:
        return 0
    return player_data['Wickets'].mean()

# Applyying the functions to create separate metrics
# Player's recent and overall performance
data['Recent Performance Player Bowler'] = data.apply(
    lambda row: calculate_weighted_recent_form(data[data['Bowler Name'] == row['Bowler Name']], row['Match Date']), axis=1)
data['Overall Performance Player Bowler'] = data.apply(
    lambda row: calculate_overall_avg_performance(data[data['Bowler Name'] == row['Bowler Name']]), axis=1)

# Player's recent and overall performance against each opponent
data['Recent Performance Opponent Bowler'] = data.apply(
    lambda row: calculate_weighted_recent_form(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Opponent Team'] == row['Opponent Team'])], row['Match Date']), axis=1)
data['Overall Performance Opponent Bowler'] = data.apply(
    lambda row: calculate_overall_avg_performance(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Opponent Team'] == row['Opponent Team'])]), axis=1)

# Player's recent and overall performance in each stadium
data['Recent Performance Stadium Bowler'] = data.apply(
    lambda row: calculate_weighted_recent_form(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Stadium Name'] == row['Stadium Name'])], row['Match Date']), axis=1)
data['Overall Performance Stadium Bowler'] = data.apply(
    lambda row: calculate_overall_avg_performance(data[(data['Bowler Name'] == row['Bowler Name']) & (data['Stadium Name'] == row['Stadium Name'])]), axis=1)

# Filling NaN values
data.fillna({'Recent Performance Player Bowler': 0, 'Overall Performance Player Bowler': 0, 
             'Recent Performance Opponent Bowler': 0, 'Overall Performance Opponent Bowler': 0, 
             'Recent Performance Stadium Bowler': 0, 'Overall Performance Stadium Bowler': 0}, inplace=True)

# Saving the preprocessed data
data.to_csv('Preprocessed_Bowler_IPL2023_Data.csv', index=False)
print(f"Preprocessed data saved as {'Preprocessed_Bowler_IPL2023_Data'}")
```

```python
import pandas as pd
from dash import Dash, dcc, html, Input, Output, State, dash_table
from pycaret.regression import load_model, predict_model

# Loading the preprocessed 2023 data for prediction
data_2023 = pd.read_csv('Preprocessed_Bowler_IPL2023_Data.csv')

# Loading the trained model
model = load_model('final_bowler_prediction_model')

def calculate_weighted_average(data, column):
    if data.empty:
        return None
    values = data[column].tolist()
    weights = sorted(values, reverse=True)
    total_weight = sum(weights)
    if total_weight == 0:
        return 0  # Return 0 to avoid division by zero
    return sum(v * w for v, w in zip(values, weights)) / total_weight

def calculate_combined_weighted_average(opponent_data, overall_data, column, weight_opponent=0.4):
    weighted_opponent = calculate_weighted_average(opponent_data, column)
    weighted_overall = calculate_weighted_average(overall_data, column)

    if weighted_opponent is None:
        weighted_opponent = 0
    if weighted_overall is None:
        weighted_overall = 0

    return (weight_opponent * weighted_opponent) + ((1 - weight_opponent) * weighted_overall)

# Initializing Dash app
app = Dash(__name__, suppress_callback_exceptions=True)

# App layout
app.layout = html.Div([
    html.H1("Cricket Match Prediction - Bowlers"),
    dcc.Dropdown(id='team1-dropdown', options=[{'label': team, 'value': team} for team in data_2023['Innings'].unique()], placeholder='Select Team 1'),
    dcc.Dropdown(id='team2-dropdown', options=[{'label': team, 'value': team} for team in data_2023['Innings'].unique()], placeholder='Select Team 2'),
    dcc.Dropdown(id='stadium-dropdown', options=[{'label': stadium, 'value': stadium} for stadium in data_2023['Stadium Name'].unique()], placeholder='Select Stadium'),
    html.Button('Submit Teams', id='submit-teams', n_clicks=0),
    html.Div(id='team-selection-state', style={'display': 'none'}),
    html.Div(id='player-selection'),
    html.Div(id='prediction-results')
])

# Callback for team and stadium selection
@app.callback(
    Output('player-selection', 'children'),
    [Input('submit-teams', 'n_clicks')],
    [State('team1-dropdown', 'value'),
     State('team2-dropdown', 'value'),
     State('stadium-dropdown', 'value')]
)
def update_player_selection(n_clicks, team1, team2, selected_stadium):
    # Debugging
    print(f"Team and Stadium Selection - Clicks: {n_clicks}, Team1: {team1}, Team2: {team2}, Stadium: {selected_stadium}")

    if n_clicks > 0:
        # Checking if teams have previous encounters this season
        previous_encounters = data_2023[(data_2023['Innings'] == team1) & (data_2023['Opponent Team'] == team2)]
        encounter_message = "Previous encounters available this season." if not previous_encounters.empty else "No previous encounters this season."
        
        # Checking if teams have played in the selected stadium
        stadium_data_team1 = data_2023[(data_2023['Innings'] == team1) & (data_2023['Stadium Name'] == selected_stadium)]
        stadium_data_team2 = data_2023[(data_2023['Innings'] == team2) & (data_2023['Stadium Name'] == selected_stadium)]
        stadium_message = ""
        if not stadium_data_team1.empty and not stadium_data_team2.empty:
            stadium_message = "Both teams have played in this stadium."
        elif not stadium_data_team1.empty:
            stadium_message = "Only Team 1 has played in this stadium."
        elif not stadium_data_team2.empty:
            stadium_message = "Only Team 2 has played in this stadium."
        else:
            stadium_message = "Neither team has played in this stadium."

        # Player selection options
        team1_players = [{'label': player, 'value': player} for player in data_2023[data_2023['Innings'] == team1]['Bowler Name'].unique()]
        team2_players = [{'label': player, 'value': player} for player in data_2023[data_2023['Innings'] == team2]['Bowler Name'].unique()]
        
        return html.Div([
            html.Div(encounter_message),
            html.Div(stadium_message),
            html.Div([
                html.H4(f"Select bowlers for {team1}:"),
                dcc.Dropdown(id='team1-players-dropdown', options=team1_players, multi=True, placeholder='Select Bowlers for Team 1')
            ]),
            html.Div([
                html.H4(f"Select bowlers for {team2}:"),
                dcc.Dropdown(id='team2-players-dropdown', options=team2_players, multi=True, placeholder='Select Bowlers for Team 2')
            ]),
            html.Button('Submit Players', id='submit-players', n_clicks=0)
        ])
    return ''

# Callback for player selection and prediction
@app.callback(
    Output('prediction-results', 'children'),
    [Input('submit-players', 'n_clicks')],
    [State('team1-players-dropdown', 'value'),
     State('team2-players-dropdown', 'value'),
     State('team1-dropdown', 'value'),
     State('team2-dropdown', 'value'),
     State('stadium-dropdown', 'value')]
)
def generate_predictions(n_clicks, team1_players, team2_players, team1, team2, selected_stadium):
    # Debugging
    print(f"Prediction - Clicks: {n_clicks}, Team1 Players: {team1_players}, Team2 Players: {team2_players}, Stadium: {selected_stadium}")

    if n_clicks > 0:
        results = []
        for player in team1_players + team2_players:
            # Debugging
            print(f"Predicting for Player: {player}")

            # Determining opponent team and innings type
            if player in team1_players:
                opponent_team = team2
                innings_type = '1st innings'
            else:
                opponent_team = team1
                innings_type = '2nd innings'

            # Filtering player data
            player_data = data_2023[data_2023['Bowler Name'] == player]

            # Calculating recent performance against the opponent
            recent_form = player_data.sort_values(by='Match Date', ascending=False).head(5)
            recent_opponent_data = player_data[player_data['Opponent Team'] == opponent_team].sort_values(by='Match Date', ascending=False).head(5)
            weighted_wickets_opponent = calculate_weighted_average(recent_opponent_data, 'Wickets')
            weighted_wickets = calculate_weighted_average(recent_form, 'Wickets')
            # Calculating combined weighted averages for features
            combined_weighted_wickets = calculate_combined_weighted_average(recent_opponent_data, recent_form, 'Wickets')
            combined_weighted_balls_bowled = calculate_combined_weighted_average(recent_opponent_data, recent_form, 'Balls Bowled')
            combined_weighted_overs = calculate_combined_weighted_average(recent_opponent_data, recent_form, 'Overs')
            combined_weighted_maidens = calculate_combined_weighted_average(recent_opponent_data, recent_form, 'Maidens')
            combined_weighted_runs_given = calculate_combined_weighted_average(recent_opponent_data, recent_form, 'Runs Given')
            combined_weighted_economy = calculate_combined_weighted_average(recent_opponent_data, recent_form, 'Economy')

            # Calculating recent performance in the stadium
            recent_stadium_data = player_data[player_data['Stadium Name'] == selected_stadium].sort_values(by='Match Date', ascending=False).head(5)
            weighted_wickets_stadium = calculate_weighted_average(recent_stadium_data, 'Wickets')

            # Calculating overall performance against the opponent
            overall_opponent_data = player_data[player_data['Opponent Team'] == opponent_team]
            avg_wickets_opponent = overall_opponent_data['Wickets'].mean() if not overall_opponent_data.empty else None

            # Calculating overall performance in the stadium
            overall_stadium_data = player_data[player_data['Stadium Name'] == selected_stadium]
            avg_wickets_stadium = overall_stadium_data['Wickets'].mean() if not overall_stadium_data.empty else None

            # Calculating overall performance of the player
            avg_wickets_overall = player_data['Wickets'].mean() if not player_data.empty else None

            # Combining all features into a single DataFrame row for prediction
            prediction_input = pd.DataFrame([{
                'Bowler Name': player,
                'Balls Bowled': combined_weighted_balls_bowled,
                'Wickets': combined_weighted_wickets,
                'Overs': combined_weighted_overs,
                'Maidens': combined_weighted_maidens,
                'Runs Given': combined_weighted_runs_given,
                'Economy': combined_weighted_economy,
                'Stadium Name': selected_stadium,
                'Opponent Team': opponent_team,
                'Innings Type': innings_type,
                'Recent Performance Player Bowler': weighted_wickets,
                'Recent Performance Opponent Bowler': weighted_wickets_opponent,
                'Recent Performance Stadium Bowler': weighted_wickets_stadium,
                'Overall Performance Opponent Bowler': avg_wickets_opponent,
                'Overall Performance Stadium Bowler': avg_wickets_stadium,
                'Overall Performance Player Bowler': avg_wickets_overall,
            }])

            # Predicting using the model
            prediction_output = predict_model(model, data=prediction_input)
            dream11_points = prediction_output['prediction_label'].iloc[0] if not prediction_output.empty else None

            # Adding to results
            prediction_input_values = prediction_input.iloc[0].to_dict()
            prediction_input_values['Player'] = player
            prediction_input_values['Predicted Dream11 Points'] = dream11_points
            results.append(prediction_input_values)

        # Creating DataFrame from results
        results_df = pd.DataFrame(results)

        # Reordering DataFrame columns to have 'Player' and 'Predicted Dream11 Points' first
        cols = ['Player', 'Predicted Dream11 Points'] + [col for col in results_df.columns if col not in ['Player', 'Predicted Dream11 Points']]
        results_df = results_df[cols]

        # Sorting DataFrame by Dream11 Points
        results_df = results_df.sort_values(by='Predicted Dream11 Points', ascending=False)

        # Displaying results in a structured table
        return dash_table.DataTable(
            data=results_df.to_dict('records'),
            columns=[{'name': i, 'id': i} for i in results_df.columns],
            style_table={'overflowX': 'auto'},
            sort_action='native',
            sort_mode='multi'
        )
    return ''

# Main execution
if __name__ == '__main__':
    app.run_server(debug=True)
```
